```python
import random
def choose_theme():
    print("По техническим причинам работает только латиница.")
    print("Есть 3 темы:\n 1. Столицы стран Южной Америки \n 2. Преподаватели Школы Лингвистики \n 3. Современные западные рэперы")
    print("Для выбора темы Столицы стран Южной Америки нажмите 1")
    print("Для выбора темы Преподаватели Школы Лингвистики нажмите 2")
    print("Для выбора темы Современные рэперы нажмите 3")
    
    #Внимание! Вводите только латиницу, кириллица не работает.
    #У меня пробел тоже считается за символ, и его надо угадывать
    #К сожалению, она не рисует человечка и виселицу(а остальное она умеет).

    

    varik = input()
    if varik == "1":
        file_name = "south_america"
    if varik == "2":
        file_name = "fikl_prepody"
    if varik == "3":
        file_name = "modern_rap"
    else:
        print("Введите только 1, 2 или 3!")
    
    

    return file_name


file_name = choose_theme()

def openfile(file_name, encoding = "utf-8"):
    file_name1 = file_name + ".txt"
    with open(file_name1) as f:
        text = f.read()
        splited_text = text.split("\n")
    return splited_text


splited_text = openfile(file_name)


def choose_word(splited_text):
    word = random.choice(splited_text)
    word_2 = list(word)
    dlina = len(word)
    slovo = ["_ " for x in word_2] 
    print("У вас есть 10 попыток, чтобы угадать слово из", dlina, "букв")
    oshibka = 0
    
    vyvod = str()
    for bukovka in slovo:
        vyvod = vyvod + bukovka
    print (vyvod)
    
    while oshibka < 10:
        naslos = 0
        bukva = input()
        l = 0
        for a in word_2:
            
            if bukva == a:
                slovo[l] = word_2[l]
                naslos = 1
                
            l = l + 1
        vyvod = ""    
        for bukovka in slovo:
            vyvod = vyvod + bukovka
        print (vyvod)
        
        popytka = 9 - oshibka
        print("У вас осталось")
        if popytka == 2 or popytka == 3 or popytka == 4:
            print(popytka, "попытки")
        if popytka == 1:
            print(popytka, "попытка")
        if popytka >= 5:
            print(popytka, "попыток")
        if naslos == 0:
            oshibka = oshibka + 1
        if popytka == 0 and slovo != word_2:
            print("Вы проиграли!")
            break
        if slovo == word_2:
            print("Вы выиграли!")
            break
    
    
choose_word(splited_text)
```
