```python
import sqlite3
import csv

with open('nanai.csv', 'r') as csv_file2:
        reader = csv.reader(csv_file2)
        spisok = list(reader)

    
header = spisok[0]
header1 = header[0]
headernew = header1.split(";")
print(header)
conn = sqlite3.connect('zadanie.db')
c = conn.cursor()
c.execute("CREATE TABLE IF NOT EXISTS vowels({}, {}, {}, {}, {}, {})".format(headernew[0], headernew[1], headernew[2], headernew[3], headernew[4], headernew[5]))

for row in spisok:
    row1 = row[0]
    rownew = row1.split(";")
    print(rownew)
    c.execute("INSERT INTO vowels VALUES (?, ?, ?, ?, ?, ?)", (rownew[0], rownew[1], rownew[2], rownew[3], rownew[4], rownew[5]))
conn.commit()
conn.close()

```
